```python
# Process to extract data and store by simulation numbers
import re
def extract_header(line):
    "Extract column header for dataframe creation"
    tokens = line.split("\t")
    #print (tokens)
    header = list()
    for key in tokens[:-1]:
        header.append(key)
    return header

def extract_data(line):
    "Extract data points by simulation"
    data_points = list()
    for i in range(len(line)-1):
        data_points.append(int(line[i]))
    return data_points

with open("plantCellSignaling.data") as f:
    lines = f.readlines()
    sim_data = list()
    count = 0
    for line in lines:
        if '*' in line:
            pass
        elif re.match('[a-zA-Z0-9]+$',line) is None:
            header = extract_header(line)
        else:
            sim_data.append(extract_data(line))
```

```python
# Helper function to generate the a list of important feature(s) by observable
def shortlisted(lst, threshold):
    """ extract node pairs that have highest significance """
    shortlisted_nodes = list()
    for predictor, importance in lst:
        if importance > threshold:
            shortlisted_nodes.append((predictor, round(importance,4)))
    return shortlisted_nodes
            
    print (predictor, round(importance,4))

def feature_importance_by_observable_rfc(df, predictors, threshold=0.1):
    lst = dict()
    for feature in predictors:
        if feature == 'ABA' or feature == 'AGB':
            pass
        else:
            clf = RandomForestClassifier(n_jobs=6)
            y = df[feature]
            pred = list(set(predictors)-set([feature])-set(['CLOSURE']))
            x = df[pred]
            clf.fit(x,y)
            zipped = list(zip(pred, clf.feature_importances_))
            #res = sorted(zipped, key = operator.itemgetter(1)) 
            lst[feature] = shortlisted(zipped, threshold)
    return lst
```

```python
def feature_importance_by_observable_etc(df, predictors, threshold=0.1):
    lst = dict()
    for feature in predictors:
        if feature == 'ABA' or feature == 'AGB':
            pass
        else:
            clf = ExtraTreesClassifier(n_jobs=6)
            y = df[feature]
            pred = list(set(predictors)-set([feature])-set(['CLOSURE']))
            x = df[pred]
            clf.fit(x,y)
            zipped = list(zip(pred, clf.feature_importances_))
            #res = sorted(zipped, key = operator.itemgetter(1)) 
            lst[feature] = shortlisted(zipped, threshold)
    return lst
```

```python
# Identify if two methods will give different inference.
def compare_nodes(lst1, lst2):
    """ helper function to compare items in two lists are same """
    set_lst1 = set([item[0] for item in lst1])
    set_lst2 = set([item[0] for item in lst2])
    if set_lst1==set_lst2:
        return True
    return False

def missing_node(lst1, lst2):
    """ helper function to return the difference """
    set_lst1 = set([item[0] for item in lst1])
    set_lst2 = set([item[0] for item in lst2])
    if len(set_lst1)>len(set_lst2):
        return list(set_lst1-set_lst2)
    else:
        return list(set_lst2-set_lst1)

for key in header:
    if key == 'ABA' or key=='AGB':
        pass
    elif not compare_nodes(node_pairs_etc[key], node_pairs_rfc[key]):
        print ('Different node(s) are identified for %s and the missing node is %s' 
               %(key, missing_node(node_pairs_etc[key], node_pairs_rfc[key])))
```
